```python
# If you're running this "locally" (on your computer, outside of this lecture and not on Binder),
# you will need to install these libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

## Histplot
#### `kde()` determines the lineplot;
#### `rug` determines the `rugplot()` on the bottom
#### `hist` determines the histplot

```python
x = np.random.normal(size=200)
sns.distplot(x,hist=True,kde=True, rug=True)
```

### We can use `kdeplot()` to draw the kde plot 
#### `shade` controls the shades
#### `bw` controls the bindwidth; similar to the `bins` in histplot

```python
sns.kdeplot(x,shade=True,bw=.1,label='bw=0.1')
sns.keplot(x,shade=True,bw=1,label='bw=1')
```

## Scatterplot

```python

data=pd.DataFrame({'x':np.random.normal(size=50),'y':np.random.normal(size=50)})
sns.jointplot(x=np.random.normal(size=100),y=np.random.normal(size=100));
sns.jointplot(x='x',y='y',data=data);
```

## Hexbin plot
#### `kind = 'hex'` change the scatterplot to hexbin plot

```python
mean, covariance = [0, 1], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, covariance, 50).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="w");
```

## Heatmap

```python
sns.heatmap([x,y])
```

## pairplot

```python
data = sns.load_dataset("iris")
sns.pairplot(data);
```

## Pairgrid

```python
data_grid = sns.PairGrid(data)
data_grid.map_diag(sns.kdeplot)
data_grid.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```
